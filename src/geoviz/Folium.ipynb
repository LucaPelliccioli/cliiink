# Essai pour la visualisation géographiques avec Folium
Pour un affichage interactif, merci de se référer à ce [lien](https://nbviewer.jupyter.org/github/antjacquemin/cliiink/blob/main/src/geoviz/Folium.ipynb) 

```python
import folium
import json
import requests
```

```python
# Création de la carte
carteCannes = folium.Map(location=[43.551062, 7.013223],zoom_start=20)

# Affichage de la pop-up indiquant latitude et longitude lorsqu'on clique sur la carte
carteCannes.add_child(folium.LatLngPopup())

# Ajout de repères avec des couleurs et logos personnélisés
folium.Marker(
    location=[43.5515, 7.0141],
    popup='Spot',
    icon=folium.Icon(icon="cloud")
).add_to(carteCannes)

folium.Marker(
    location=[43.5510, 7.0181],
    popup='Palais des Festivals',
    icon=folium.Icon(color='green')
).add_to(carteCannes)

folium.Marker(
    location=[43.5517, 7.0142],
    popup='un repère au hasard',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(carteCannes)

# Ajout d'un repère avec affichage de graphe (utile pour l'évolution de l'utilsation d'un collecteur)
stats = json.loads(requests.get("https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/vis1.json").text)

folium.Marker(
    location=[43.5512, 7.0123],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(stats, width=450, height=250))
).add_to(carteCannes)

# Ajout d'un cercle
folium.CircleMarker(
    location=[43.5514, 7.0126],
    radius=30,
    popup='Hôtel de ville de Cannes',
    color='red',
    fill=True,
    fill_color='crimson'
).add_to(carteCannes)

carteCannes
```

```python
# Adresse du fichier
fichierCACPL = "../../data/cacpl.geojson"

# Conversion du fichier (Geo)JSON en dictionnaire
with open(fichierCACPL) as fichier:
    dictCACPL = json.load(fichier)

carteCACPL = folium.Map(location=[43.551062, 7.013223],zoom_start=13)

# Parcours du dcitionnaire pour générer chaque polygone interactif de chaque ville
for feature in dictCACPL["features"]:
    gj = folium.GeoJson(
        data=feature["geometry"]
    )
    gj.add_child(folium.Popup(feature["properties"]["nom"]))
    gj.add_to(carteCACPL)


carteCACPL
```
